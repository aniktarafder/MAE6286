```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
T = 100.0
dt= 0.02
N = int(T/dt)+1
t = numpy.linspace(0.0, T, N)
```

```python
t = numpy.arange(0.0, 100.02, 0.02)
```

```python
#initial values
z0 = 100.0
b0 = numpy.array([0.0, 1.0, 100.0])
zt = 100.0
g = 9.81
numpy.size(b0)
b0
```

```python
for i in range(numpy.size(b0)):
    u = numpy.array([z0, b0[i]])
    z = numpy.zeros(N)
    z[0] = z0
    for n in range(1, N):
        rhs = numpy.array([u[1], g * (1 - u[0] / zt)])
        u = u + dt * rhs
        z[n] = u[0]
        
        pyplot.figure(figsize=(10.0, 5.0))
        pyplot.title('Elevation of the phugoid over the time') 
        pyplot.ylabel('Elevation', fontsize = 16)
        pyplot.xlabel('time(s)', fontsize = 16)
        pyplot.xlim(60, 140)
        pyplot.xlim(0.0, t[5000])
        pyplot.plot(t,z, color = 'b', linewidth = 2)
```

```python
pyplot.figure(figsize=(10.0, 5.0))
pyplot.title('Elevation of the phugoid over the time') 
pyplot.ylabel('Elevation', fontsize = 16)
pyplot.xlabel('time(s)', fontsize = 16)
pyplot.xlim(60, 140)
pyplot.xlim(0.0, t[5000])
pyplot.plot(t,z, color = 'b', linewidth = 2)
```
