```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
nx =41
L = 2.0
dx = L/(nx-1)
nt = 25
dt = 0.01
c = 1.0
x = numpy.linspace(0.0, L, num=nx)
```

```python
u0 = numpy.ones(nx)
mask = numpy.where(numpy.logical_and(x >=0.5, x<=1.0))
mask
```

```python
u0[mask] = 2.0
```

```python
def plot_figure(x,u0, u):
    pyplot.figure(figsize=(4.0,4.0))
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.grid()
    pyplot.plot(x, u0, label='initial')
    pyplot.plot(x, u, label='current')
    pyplot.legend()
    pyplot.xlim(0.0,2.0)
    pyplot.ylim(0.0,2.5)
```

```python
plot_figure(x,u0,u0)
```

```python
u = u0.copy()
for n in range(1,nt):
    un = u.copy()
    for i in range (1, nx):
        u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])
```

```python
plot_figure(x,u0,u)
```

# 1D non-linear convection

```python
u = u0.copy()
for n in range(1,nt):
    un = u.copy()
    for i in range (1, nx):
        u[i] = un[i] - u[i] * dt / dx * (un[i] - un[i-1])
```

```python
plot_figure(x,u0,u)
```

```python
u = u0.copy()
for n in range(1,nt):
    un = u.copy()
#    for i in range (1, nx):
#        u[i] = un[i] - u[i] * dt / dx * (un[i] - un[i-1])
    u[1:] = un[1:] - u[1:] * dt /dx * (un[1:]-un[:-1])
```

```python
plot_figure(x,u0,u)
```
